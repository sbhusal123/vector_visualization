# Divergence Of  A Vector Field

Divergence is applied to a point on a vector field to check weather the point on a vector field is a source, sink or neither of it.

Consider example of a water filling on a gallon through a household tap, a tap can be considered as a souce, a  gallon can be considered as a sink and a flowing stream of a water can be considered as a points where it's neither sink nor source, just s stream of water. It is denoted as $\nabla$.

The operator $\nabla$ is equivalent to:

$$
\nabla = \frac{\partial}{\partial{x}} + \frac{\partial}{\partial{y}} + \frac{\partial}{\partial{z}}
$$

Consider a vector field

$$
\vec{F_{(x, y, z)}} = F_x \hat{i} + F_y \hat{j} + F_z \hat{k}
$$

Then the divergence of a vector field is given as:

$$
\nabla . \vec{F} = \frac{\partial{F_x}}{\partial{x}} + \frac{\partial{F_y}}{\partial{y}} + \frac{\partial{F_z}}{\partial{z}}
$$

Note, we're using a **dot prodct of a divergence operator and our vector.**

# Example 1:

Given a vector field

$$
\vec{F_{(x, y)}} = sin(x) \hat{i} + cos(y) \hat{j}
$$

Plot the vector field and interpret the plot.

**With this example, we'll see where the point acts as a source, sink.**

Let's start by doing hands on.

$$
\nabla . \vec{F_{(x,y)}} = \frac{\partial{sin(x)}}{\partial{x}} + \frac{\partial{cos(y)}}{\partial{y}}
$$

Which can be resolved to:

$$
\nabla . \vec{F_{(x,y)}} = cos(x) - sin(y)
$$

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We're constructing a grid of points accross x and y staring from 0 to pie. (180) with 20 points.

Note, that we're taking a radian equivalent angle.i.e. 

- 0 = 0
- 180 = pi

```python
x = np.linspace(0, np.pi, 20)
y = np.linspace(0, np.pi, 20)
```

Forming a meshgrid

```python
x, y = np.meshgrid(x, y)
```

Calculating scalar quantity of our basis vectors i and j

```python
u = np.sin(x)
v = np.cos(y)
```

Plotting our field

```python
# fig, ax = plt.subplots(figsize=(50, 50))
# fig.dpi = 30
plt.quiver(x, y, u, v, np.hypot(u, v), cmap=plt.cm.gist_heat_r, scale=25,angles='uv')
plt.xlabel("x")
plt.ylabel("Y")
plt.show()
```

Note: though the x is labled from 0 to 3 (in terms of p), the values represents a numeric values of pi from 0 to pi.

# Conclusion

- We can clearly see that at point (0,0) in the above plot, the arrows are emerging and it acts as a source.
- At the top right point (pi, pi) it acts as a sink.
