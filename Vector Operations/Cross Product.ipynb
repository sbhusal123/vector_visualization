## 1. Cross Product

Condider the two vector a and b.

$$
\vec{a} = a_{x} \hat{i} + a_{y} \hat{j} + a_{z} \hat{k}
$$

$$
\vec{b} = b_{x} \hat{i} + b_{y} \hat{j} + b_{z} \hat{k}
$$

Then the cross product: $\vec{a} X \vec{b}$ can be computed as:

\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
a_x & a_y & a_z \\
b_x & b_y & b_z \\
\end{vmatrix}

## Example. Plot the cross product of the following two vector.

$$
\vec{a} = x \hat{i} + y \hat{j} + z \hat{k}
$$

$$
\vec{b} = x \hat{i} + y \hat{j} - z \hat{k}
$$

Setting up our grid for vector a

````python
# for field a
a_x, a_y, a_z = np.meshgrid(
    np.linspace(1, 5, 5),
    np.linspace(1, 5, 5),
    np.linspace(1, 5, 5)
)
# magintude of basis
a_u  = 1
a_v = 1
a_w = 1

# for field b
b_x, b_y,b_z = np.meshgrid(
    np.linspace(1, 5, 5),
    np.linspace(1, 5, 5),
    np.linspace(1, 5, 5)
)

# magintude of basis
b_u = 1
b_v = 1
b_w = - 1
````

## Configure 3d subplots- 1 for vector a

````python
fig = plt.figure(figsize=(15, 30))
ax1 = fig.add_subplot(3, 1, 1, projection='3d')
````

Setting up Limits For Our Plots

````python
ax1.axes.set_xlim3d(left=0, right=6) 
ax1.axes.set_ylim3d(bottom=0, top=6) 
ax1.axes.set_zlim3d(bottom=0, top=6)
````

````python
ax1.set_xlabel('X', fontsize=20, rotation = 0)
ax1.set_ylabel('Y', fontsize=20, rotation = 0)
ax1.set_zlabel('Z', fontsize=20, rotation = 0)
````

Quiver Vector Field Plot

````python
ax1.quiver(a_x, a_y, a_z, a_u, a_v, a_w, length=1, normalize=True)
plt.show()
````

## Plotting the cross product

We've two vector fields as:

$$
\vec{a} = x \hat{i} + y \hat{j} + z \hat{k}
$$

$$
\vec{b} = x \hat{i} + y \hat{j} - z \hat{k}
$$

To calculate their cross product we've:



\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
x & y & z \\
x & y & -z \\
\end{vmatrix}

Which yields us with following equation as:

$$
\vec{a}X\vec{b}=-2yz \hat{i} + 2xz \hat{j}
$$

````python
# for field a
ab_x, ab_y, ab_z = np.meshgrid(
    np.linspace(1, 5, 5),
    np.linspace(1, 5, 5),
    np.linspace(1, 5, 5)
)
ab_u  = -2*ab_y*ab_z
ab_v = 2*ab_x*ab_z
ab_w = 0
````

````python
ax3 = fig.add_subplot(3, 1, 3, projection='3d')

ax3.axes.set_xlim3d(left=0, right=6) 
ax3.axes.set_ylim3d(bottom=0, top=6) 
ax3.axes.set_zlim3d(bottom=0, top=6)


ax3.set_xlabel('X', fontsize=20, rotation = 0)
ax3.set_ylabel('Y', fontsize=20, rotation = 0)
ax3.set_zlabel('Z', fontsize=20, rotation = 0)

````

````python
ax3.quiver(ab_x, ab_y, ab_z, ab_u, ab_v, ab_w, length=1, normalize=True, color='r')
plt.show()
````
